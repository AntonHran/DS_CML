```python
import random
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


def hypothesis(theta, x):
    return np.dot(theta, x)
```

```python
def mean_squared_error_(y_true: np.array, y_pred: np.array) -> float:
    return np.mean((y_true - y_pred)**2)/2
```

```python
matr: np.array = np.array([[1, 1], [1, 1]])  # for example
w = np.array([random. uniform(-5, 5) for i in range(matr.shape[1] + 1)])  # starting coeffs for calculating

def one_step_gradient(x: np.array, y: np.array, w: np.array, lr: float = 0.001): # realization see below
    m = len(y)
    h = np.dot(w, x.transpose())
    w = w - lr * np.dot(x.transpose(), (h - y)) / m
    return w
```

```python
path = "Housing.csv"
data = pd.read_csv(path)
data
```

```python
columns = ["area", "bedrooms", "bathrooms"]
data_new = data[columns]
```

```python
m = np.array(data_new)
m
```

```python
matr_X = data_new.values

print(len(matr_X), matr_X.shape)
```

```python
matr_y = np.array(data["price"])
print(len(matr_y))
```

```python
matr_X[109:]
```

```python
w = np.array([random. uniform(-5, 5) for i in range(matr_X.shape[1] + 1)])
w
```

```python
add_matr = np.ones((len(matr_X), 1))
X = np.concatenate((add_matr, matr_X), axis=1)
res_one = one_step_gradient(X, matr_y, w)
res_one
```

```python
model = LinearRegression()
regressor_task = model.fit(matr_X[109:], matr_y[109:])

print(regressor_task.coef_)
print(regressor_task.intercept_)

y_pred_test = model.predict(matr_X[:109])
print(y_pred_test)
print("Mean squared error: %.2f" % mean_squared_error(matr_y[:109], y_pred_test))
print(regressor_task.predict(np.array([[7420, 4, 2]])))
# [1.40146963e+06 2.39630763e+02 3.01147193e+05 5.33804306e+05] - from analytical resolve
```
